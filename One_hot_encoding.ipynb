```python
import pandas as pd

def load_data():
  return pd.read_csv('https://raw.githubusercontent.com/BindiChen/machine-learning/main/data-analysis/001-pandad-pipe-function/data/train.csv'
  )

df = load_data()
df.head()
```

# 1. Creating a One-Hot encoding

```python
pd.get_dummies(df.Sex)
```

# Creating a Dummy encoding

```python
pd.get_dummies(df.Sex, drop_first=True)
```

# 3. Appending prefix to column names

```python
pd.get_dummies(df.Sex, prefix='Sex')
```

# 4. For a variable with more than two possible values

```python
df.Embarked.value_counts()
```

```python
pd.get_dummies(df.Embarked, prefix='Embarked')
```

# 5. Attach dummies to the dataframe

```python
embarked_dummies = pd.get_dummies(df.Embarked, prefix='Embarked')
df = pd.concat([df, embarked_dummies], axis=1)
```

```python
df.head()
```

# 6.Pass a dataframe

```python
df = load_data() # reset dataframe
df = pd.get_dummies(df, columns=['Sex','Embarked'])
```
